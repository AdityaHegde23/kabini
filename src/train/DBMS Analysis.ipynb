```python
import pandas as pd
import numpy as np
import itertools
from scipy.stats import expon

csv_file = "your_data_distribution.csv"
data = data_dist#pd.read_csv(csv_file)
numeric_attributes = data.columns.tolist()
random_queries_count = 200
vector_data_columns = [
    f"{col}_{suffix}" for col in numeric_attributes for suffix in ["start", "end"]
]
vector_data_randq = [(0,) * len(vector_data_columns)] * random_queries_count
vector_df_randq = pd.DataFrame(vector_data_randq, columns=vector_data_columns)



# Function to generate query centers and widths
def generate_query_parameters(attr_min, attr_max, distribution="uniform"):
    if distribution == "uniform":
        center = np.random.randint(attr_min, attr_max + 1)
        width = np.random.randint(1, attr_max - attr_min + 1)
    else:  # Exponential distribution for data-centric
        scale = (attr_max - attr_min) / 3
        width = int(expon(scale=scale).rvs())
        center = np.random.randint(attr_min, attr_max + 1)
    return center, width

# Generate SQL queries
queries = []

for row_number in range(random_queries_count):
    # Randomly select the number of dimensions (attributes) for the query
    num_attrs = np.random.randint(2, len(numeric_attributes) + 1)
    selected_attrs = np.random.choice(numeric_attributes, size=num_attrs, replace=False)

    conditions = []
    for attr in selected_attrs:
        attr_min, attr_max = (
            data[attr].min(),
            data[attr].max(),
        )  # Assuming min and max values are representative
        center, width = generate_query_parameters(
            attr_min, attr_max, distribution="uniform"
        )
        min_val = max(attr_min, center - width // 2)
        max_val = min(attr_max, center + width // 2)
        conditions.append(f"'{attr}' BETWEEN {min_val} AND {max_val}")
        vector_df_randq.at[row_number, f"{attr}_start"] = min_val
        vector_df_randq.at[row_number, f"{attr}_end"] = max_val

    query = "SELECT * FROM your_table WHERE " + " AND ".join(conditions) + ";"
    queries.append(query)

# Optionally, save the queries to a file
with open("generated_queries.sql", "w") as f:
    for query in queries:
        print(query)
        f.write("%s\n" % query)

print("Queries generated and saved to generated_queries.sql")
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

def load_data(data, train_size=0.8):

  labels_path = f'/content/drive/MyDrive/CS236 DBMS/Project/{data}_labels.csv'
  features_path = f'/content/drive/MyDrive/CS236 DBMS/Project/{data}_vector.csv'

  labels_data = pd.read_csv(labels_path)
  features_data = pd.read_csv(features_path, header=None)


  labels = labels_data['actual_value'].values

  # To handle zero actual selectivity values
  labels += 1


  labels = np.log2(labels)

  # Normalizing features since neural networks perform best on this
  # this step helped to reduce the loss and get better predicted selectivity values
  features_array = features_data.values
  scaler = StandardScaler()
  features_array = scaler.fit_transform(features_array)


  X_train, X_test, y_train, y_test = train_test_split(features_array, labels, train_size = train_size, random_state = 42)

  return X_train, X_test, y_train, y_test
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
import torch.optim as optim

from sklearn.metrics import mean_absolute_error
```

```python
class QueriesDataset(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels

    def __len__(self):
        return len(self.features)

    def __getitem__(self, index):
        return self.features[index], self.labels[index]
```

```python
class SelectivityMLP(nn.Module):
    def __init__(self, dim):

        super(SelectivityMLP, self).__init__()

        self.layers = nn.Sequential(
            nn.Linear(dim, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 1)
        )

    def forward(self, x):
        return self.layers(x)
```

```python
def train(dim, train_dataset, val_dataset, epochs = 500):

    model = SelectivityMLP(dim = dim) # No of attributes x 2 (upper bound, lower bound for each attribute)
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    criterion = nn.MSELoss()

    total_val_loss = 0


    train_loader = DataLoader(train_dataset, batch_size = 32, shuffle = True)
    val_loader = DataLoader(val_dataset, batch_size = 32, shuffle = True)

    L = len(train_loader)

    for epoch in range(epochs):

        model.train()
        total_loss = 0
        train_pred = []
        val_pred = []
        val_actual = []
        train_actual = []

        for inputs, labels in train_loader:
          #
            optimizer.zero_grad()
            outputs = model(inputs)

            loss = criterion(outputs, labels)
            loss.backward()

            optimizer.step()

            total_loss += loss.item() / L

            train_pred.extend(list(outputs.squeeze(1).detach().numpy()))
            train_actual.extend(list(labels.squeeze(1).detach().numpy()))

            #print(train_pred)

        train_mae = mean_absolute_error(
                train_pred,
                train_actual
            )

        for inputs, labels in val_loader:
          model.eval()

          output = model(inputs)

          loss = criterion(output, labels)

          val_pred.extend(list(output.detach().numpy()))
          val_actual.extend(list(labels.detach().numpy()))

          total_val_loss += loss.item() / L

        val_mae = mean_absolute_error(
                val_pred,
                val_actual
            )

        if epoch % 50 == 0:
          print('Epoch {}: Train Loss: {:.3f} Train MAE {:.3f} VAL MAE: {:.3f}'.format(epoch, total_loss, train_mae, val_mae))

    return model
```

```python

def train_neural_net(dataset, train_size=0.8, epochs=500): #default Epochs. Change epochs here

  #Vary train size to impact the # of training samples used
  #train_size 0.8 = 16000 samples
  #train_size 0.6 = 12000 samples, etc..

  X_train, X_test, y_train, y_test = load_data(dataset, train_size)


  X_train = torch.tensor(X_train, dtype=torch.float32)
  X_test = torch.tensor(X_test, dtype=torch.float32)
  y_train = torch.tensor(y_train, dtype=torch.float32).view(-1, 1)
  y_test = torch.tensor(y_test, dtype=torch.float32).view(-1, 1)


  train_dataset = QueriesDataset(X_train, y_train)
  test_dataset = QueriesDataset(X_test, y_test)

  dim = X_train.shape[1]

  return train(dim, train_dataset, test_dataset, epochs = epochs), train_dataset, test_dataset
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

def get_q_error(estimation, actuals):
    error1 = np.nan_to_num(np.array(estimation) / np.array(actuals))
    error2 = np.nan_to_num(np.array(actuals) / np.array(estimation))
    q_error = np.maximum(error1, error2)
    return q_error

def run_svm(dataset):

  svr_model = SVR(kernel='rbf')

  X_train, X_test, y_train, y_test = load_data(dataset)


  svr_model.fit(X_train, y_train)


  predictions = svr_model.predict(X_test)


  mse = mean_squared_error(y_test, predictions)
  print(f"Mean Squared Error: {mse}")

  return svr_model, (X_train, y_train), (X_test, y_test)


def evaluate_svm(svm, dataset):

  output = svm.predict(dataset[0])
  actuals = dataset[1]

  estimation = np.exp2(output)
  actuals = np.exp2(actuals)

  q_error = get_q_error(estimation, actuals)

  return q_error

n_estimators_list = ['forest_cover', 'power', 'higgs' ]
q_errors_svm = []

for n_estimators in n_estimators_list:
    svm, train_data, test_data = run_svm(n_estimators)

    q_error = evaluate_svm(svm, test_data)
    q_errors_svm.append(q_error)
```

```python
names = ['Forest_Cover', 'Power', 'Higgs']

plt.figure(figsize=(10, 4))
plt.boxplot(q_errors_svm, labels=names, showfliers=False)
plt.yscale('log')  # Setting the y-axis to logarithmic scale
plt.xlabel('Dataset')
plt.ylabel('Q_Error')
plt.title('Box Plot of SVM')
plt.show()
```

```python
import matplotlib.pyplot as plt

def evaluate_neural_network(model, dataset):

    dataloader = DataLoader(dataset, batch_size = 32, shuffle = True)
    model.eval()

    predictions, actuals = [], []

    with torch.no_grad():

        for inputs, labels in dataloader:

            outputs = model(inputs)

            predictions.extend(outputs.view(-1).tolist())
            actuals.extend(labels.view(-1).tolist())


    predictions = np.exp2(predictions)
    actuals = np.exp2(actuals)

    error1 = np.nan_to_num(np.array(predictions) / np.array(actuals))
    error2 = np.nan_to_num(np.array(actuals) / np.array(predictions))
    q_error = np.maximum(error1, error2)
    return q_error

q_errors_nn = []

for i in range(3):
  q_error = evaluate_neural_network(models[i], test_datasets[i])
  q_errors_nn.append(q_error)

names = ['Forest_Cover', 'Power', 'Higgs']

plt.figure(figsize=(10, 4))
plt.boxplot(q_errors_nn, labels=names, showfliers=False)
plt.yscale('log')  # Setting the y-axis to logarithmic scale
plt.xlabel('Dataset')
plt.ylabel('Q_Error')
plt.title('Box Plot of NN')
plt.show()
```
